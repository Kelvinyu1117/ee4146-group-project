```python
import pickle as pk
```

```python
BASE_PATH = '../data/'
TRAIN_FILE = 'train_processed.pkl'
EVALUATION_FILE = 'test_processed.pkl'    
```

```python
with open(BASE_PATH + TRAIN_FILE, 'rb') as f:
    train = pk.load(f)
```

```python
train
```

```python
X = train.loc[:, train.columns != 'Category']
Y = train.loc[:, train.columns == 'Category']
```

```python
X
```

```python
Y
```

```python
# Cross validation: Random forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=100, random_state=0, n_jobs=4, verbose=1)
```

```python
import time
from sklearn.model_selection import cross_validate

t1 = time.time()
cv_results = cross_validate(clf, X, Y.values.ravel(), cv=3)
t2 = time.time()
print(f'Total time: {t2 - t1} s, {int(t2 - t1) / 60} mins')
```
